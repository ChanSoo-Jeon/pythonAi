```python
import numpy as np
import matplotlib.pyplot as plt


step = lambda x : x> 0.00001

x = np.arange(-10.0,10.0,0.1)
y =  step(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

relu = lambda x : np.maximum(x,0)

x = np.arange(-10.0,10.0,0.1)
y = relu(x)
plt.plot(x,y)
plt.grid()
plt.show()
```
