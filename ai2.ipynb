```python
#초기값 조건문 증감값
cnt = 0 #기초값

while cnt <3: #조건문 = stop condition -> 최적화 optimization
  print("Hello World")
  cnt = cnt+1 #증감값
```

```python
cnt = 0
flag = False  #상태변수 : 로직을 돌리고 있을때 멈출 수 있는 동작을 담당하는 변수 -> state machine

while cnt <100:
  if cnt >2:
    flag = True
  if flag == True:
    break
  print("Hello World")
  cnt = cnt+1
```

```python
import numpy as np
import math

sum = 0 #합계
ave = 0 #평균
std = 0 #표준편차
vsum = 0  #분산
nummin = 0 #최소값
nummax = 0 #최대값
mid = 0 #중간값

n = list(range(1,101))

for i in n :
  sum += i
ave = sum/len(n)
vsum = np.var(n)
std = math.sqrt(vsum)
nummin = min(n)
nummax = max(n)
mid = np.median(n)

print(sum)
print(ave)
print(vsum)
print(std)
print(nummin)
print(nummax)
print(mid)
```

```python
print("Input your name :",end="")
name1 = input()

print("Input your age :",end="")
age1 = int(input())

# 1. 서식지정 연산자
str1 = "Your name is %s and you are %d years old"%\
(name1,age1)
print(str1)

# 2. format()
str1 = "Your name is {} and you are {} years old". \
format(name1,age1)
print(str1)

# 3.f-string
str1  = f"Your nams is {name1} and you are {age1} years old"
print(str1)
```

```python
lt1 = []


def append_list():
  key = input('Input Data :')
  lt1.append(key)
  
def print_list():
  print(lt1)

def delete_list():
  key = input('Input Data :')
  while key in lt1:
    lt1.remove(key)
    if key not in lt1:
      pass
  

while True:
  print('1. Append List')
  print('2. Print List')
  print('3. Delete List')
  print('4. Exit')

  key = input('Select Number [1~3] :')
  if key == '1':
    append_list()
  elif key == '2':
    print_list()
  elif key == '3':
    delete_list()
  else:
    break

print('Thanks.')
```

```python
ss_in = int(input())
ss = 0
nn = 1

while (ss < ss_in):
  ss += nn
  if(ss==ss_in):
    print(ss,nn)
    break
  elif ((ss_in - ss) < (ss+(nn+1) - ss_in)):
    print(ss,nn)
    break
  nn += 1
```
