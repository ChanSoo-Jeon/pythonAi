##### 03. 대량의 데이터 세트에서 경사 하강법을 수행할 때 배치 크기는 어느 정도로 하여야 할까?

배치 크기가 너무 클 경우 학업 속도가 느려지고, 메모리 부족이 발생할 경우가 있으므로 적당한 사이즈로 만든다.

##### 04. 본문의 참고 박스에서 이야기한대로 독립 변수와 종속 변수가 각각 하나인 선형 회귀에서 손실 함수의 형태는 2차 함수로 단순하기 때문에 분석적인 방법으로도 최소값을 계산할 수 있다.


${W} = \frac{\displaystyle\sum_{i=1}^{n}{(x_i-\bar x)(y_i-\bar y)}}{\sum_{i=1}^{n}{(x_i-\bar x)^2}} , b = \bar y - {W}\bar x$

##### 위의 수식 계산을 파이썬 프로그램으로 구현해보자. 경사 하강법과 비교해보자. 차이가 있는가? 이렇게 분석적인 방법으로 계산할 수 있는데, 어떠한 경우에 경사 하강법을 사용하는가?

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([0.0,1.0,2.0])
y = np.array([3.0,3.5,5.5])

W = 0
b = 0


epochs = 1000

n = float(len(X))
#분석적인 방법
for i in range(epochs):
  W = sum((X-np.mean(X))*(y-np.mean(y))) / sum((X-np.mean(X))**2)
  b = np.mean(y) - (W*np.mean(X))

print(W,b)

y_pred = W*X + b

plt.scatter(X,y)

plt.plot([min(X),max(X)],[min(y_pred),max(y_pred)],color='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([0.0,1.0,2.0])
y = np.array([3.0,3.5,5.5])

W = 0
b = 0

lrate = 0.01
epochs = 1000

n = float(len(X))
#경사하강법
for i in range(epochs):
  y_pred = W*X + b
  dw = (2/n) * sum(X*(y_pred-y))
  db = (2/n) * sum(y_pred-y)
  W = W - lrate * dw
  b = b - lrate * db

print(W,b)

y_pred = W*X + b

plt.scatter(X,y)

plt.plot([min(X),max(X)],[min(y_pred),max(y_pred)],color='red')
plt.show()
```

##### 05. 다음 표는 어떤 회사의 연도별 매출 실적을 정리한 표이다.




x(연도) | 2015 | 2016 | 2017 | 2018 | 2019 |
---     |      |      |      |      |      |
y(억원) |  12  |  19  |  28  |  37  |  46  |


(a) 선형 회귀로 분석하여 직선의 방적식 y = mx + b를 찾아보자. 본문의 프로그램을 이용해도 좋다.  ${y = 8.6x + ? (2.6)}$

```python
import matplotlib.pylab as plt
from sklearn import linear_model
import numpy as np
from sklearn.preprocessing import MinMaxScaler

reg = linear_model.LinearRegression()

X = np.array([[2015],[2016],[2017],[2018],[2019]])
y = np.array([12,19,28,37,46])

reg.fit(X,y)


print(reg.predict([[2020]]))
print(reg.coef_)
print(reg.intercept_)

plt.scatter(X,y,color="black")
y_pred = reg.predict(X)

plt.plot(X,y_pred, color = 'blue',linewidth=3)
plt.show()
```

```python
import matplotlib.pylab as plt
from sklearn import linear_model
import numpy as np
from sklearn.preprocessing import MinMaxScaler

reg = linear_model.LinearRegression()

X = np.array([[1930],[1940],[1950],[1960],[1970],[1980],[1990],[2010],[2016]])
y = np.array([59,62,70,69,71,74,75,76,78])

reg.fit(X,y)


print(reg.predict([[1962]]))
print(reg.coef_)
print(reg.intercept_)

plt.scatter(X,y,color="black")
y_pred = reg.predict(X)

plt.plot(X,y_pred, color = 'blue',linewidth=3)
plt.show()
```

##### ${y = 2.7x + 3.7}$

```python
import matplotlib.pylab as plt
from sklearn import linear_model
import numpy as np
from sklearn.preprocessing import MinMaxScaler

reg = linear_model.LinearRegression()

X = np.array([[30],[35],[20],[15],[3]])
y = np.array([90,95,70,40,10])

reg.fit(X,y)



print(reg.coef_)
print(reg.intercept_)

plt.scatter(X,y,color="black")
y_pred = reg.predict(X)

plt.plot(X,y_pred, color = 'blue',linewidth=3)
plt.show()
```
