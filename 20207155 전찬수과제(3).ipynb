```python
import numpy as np
from sklearn.metrics import mean_squared_error 
import matplotlib.pyplot as plt
import pandas as pd


xx = np.arange(0.0,10)
yy = 2.0*xx + 5
yy2 = np.random.randn(10) + yy
yy3 = abs(yy-yy2)
MSE = (np.square(yy2-yy)).mean()


combArr = {'xx':xx,'yy':yy,'yy2':yy2,'diff':yy3}
result = pd.DataFrame(data=combArr)


plt.plot(xx,yy)
plt.scatter(xx,yy2,c='orange')
plt.grid()
plt.show()

print(result)
print(f"MSE : {MSE}")
```
