1. 신경망에서는 다양한 활성화 함수가 사용된다. 어떤 활성화 함수들이 사용되는가? 각 활성화 함수들의 특징은 무엇인가?


답 : 
계단 함수: 입력 신호의 총합이 0을 넘으면 1을 출력하고 그렇지 않으면 0을 출력하는 함수.

시그모이드함수 : 0.0에서 1.0까지의 연속적인 실수를 출력하는 함수. 계단 함수와 달리 미분가능하며, 경사하강법 기법을 사용할 수 있다. 하지만 양수만 출력되어 학습 경로가 진동하면서 학습 속도가 느려진다. 함숫값이 0과 1로 수렴하기때문에 그래디언트 포화가 발생한다.

ReLU함수 : 입력이 0을 넘으면 그대로 출력하고, 입력이 0보다 작으면 출력은 0이 된다. ReLU함수는 미분이 간단하고 그래디언트 감쇠가 일어나지 않는다. 시그모이드보다 6배 정도 빠르다.

TanH 함수 : 넘파이에서 제공하며 시그모이드 함수와 비슷하지만 그 출력값이 -1에서 1까지이다. 

2. 본문에서 오차함수가 $y=x^2-6x+4$ 일 때 경사 하강법을 사용하여 오차함수 y의 

답 : 최저값을 계산하는 절차를 설명한 바 있다. 이것을 다음과 같이 그래프로 그릴 수 있는가? 맵플롯립을 사용해보자. 학습률을 변화시키면서 점을 그려보자. 어떻게 점들이 그려지는가? 학습률을 아주 크거나 작게 설정해서 점들을 그려보자.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
learning_rate = 0.4
precision = 0.00001
max_iterations = 100
x1 = []
y = []

loss_func = lambda x : x**2-6*x+4
gradient = lambda x:2*x-6

for i in range(max_iterations):
  newx =x-learning_rate*gradient(x)
  x1.append(newx)
  y.append(loss_func(newx))
  


plt.plot(x,loss_func(x))
plt.scatter(x1,y)
plt.show()
```
