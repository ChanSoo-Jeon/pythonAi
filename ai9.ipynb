```python
#M-2

lt1 = []

def append_List():
  key = input('Append Data : ')
  lt1.append(key)
  print(lt1)

def print_List():
  print(lt1)
  pass

def delete_List():
  print(lt1)
  key = input('Delete Data : ')
 
  if set(lt1) & set(key) == set():
    print('No Data')
  else :
    while True:
      lt1.remove(key)
      if set(lt1) & set(key)==set():
        break;

    print(lt1)

while True:
  print('1. Append List') 
  print('2. Print List')
  print('3. Delete List') 
  print('4. Exit') 

  key = input('Select Number [1~4] :') 
  if key == '1':
    append_List()
  elif key == '2':
    print_List()
  elif key == '3':
    delete_List()
  else:
    break

print('\nThanks.')
```

```python
#M-3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(0,9,10)
y1 = 2*x+5
y2 = y1 + 3*np.random.randn(10)

plt.plot(x,y1)
plt.scatter(x,y2)
plt.grid()
plt.show()

d = pd.DataFrame({'x' : x,'y1':y1,'y2':y2,'diff':abs(y1-y2)})
print(d)
mse = np.sum((y1-y2)**2)/10
print(f'MSE = {mse}')
```

```python
#M-4
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
  s=1/(1+np.exp(-x))
  ds = s*(1-s)
  return s,ds

x = np.linspace(-10,10,100)
y1, y2 = sigmoid(x)

plt.plot(x,y1,x,y2)
plt.xlabel("x")
plt.ylabel("Sigmoid(x), Sigmoid'(x)")
plt.show()
```

```python
#P30-0060
import numpy as np
np1 = np.random.random(3*4).reshape((3,4))

print(np1)

mu = np1.mean() #np1의 평균
sigma = np1.std() # np1의 표준편차
print(mu,' ',sigma)
wt = (np1-mu)/sigma #표준화
print(wt.round(2))

import matplotlib.pyplot as plt
plt.subplot(2,1,1)
plt.hist(np1)
plt.subplot(2,1,2)
plt.hist(wt)
plt.show()
```

```python
#P90-0062
import numpy as np

np1 = np.random.randint(0,10,10)
#np1 = (np.random.rand(10)*10).astype('int32')
#np1 = (np.random.uniform(0,10,10)).astype('int32')

print(np1)

mm = np1.min()
MM = np1.max()

mm2 = 15
MM2 = 25

wt = (np1 - mm) / (MM-mm)
np2 = wt*(MM2-mm2) + mm2
np2 = np2.astype('int64')
print(np2,type(np2))

import matplotlib.pyplot as plt
plt.subplot(2,1,1)
plt.hist(np1)
plt.subplot(2,1,2)
plt.hist(np2)
plt.show()
```
