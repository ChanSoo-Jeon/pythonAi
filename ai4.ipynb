```python
# 딥러닝 교재 98p 문제
# 시그모이드 함수 -> 종족변수 0~1
# 프로그램 해보기
```

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
  s=1/(1+np.exp(-x))
  ds = s*(1-s)
  return s,ds

x = np.linspace(-10,10,100)
y1, y2 = sigmoid(x)

plt.plot(x,y1,x,y2)
plt.xlabel("x")
plt.ylabel("Sigmoid(x), Sigmoid'(x)")
plt.show()
```
