```python
#문제 P00-0024
str1 = input()
print(str1)

def myshift_left(str):
  str = list(str)
  temp = str[0]
  for i in range(1,len(str)):
    str[i-1] = str[i]

  str[len(str)-1] = temp
  str = ''.join(str)
  return str

print(myshift_left(str1))
```

```python
#문제 P00-0025
str1 = input()
print(str1)

def myshift_right(str):
  str = list(str)
  temp = str[len(str)-1]
  for i in range(1,len(str)):
    str[-i] = str[-i-1]
  str[0] = temp
  str = ''.join(str)
  return str

print(myshift_right(str1))
```
