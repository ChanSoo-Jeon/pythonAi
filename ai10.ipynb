```python
import numpy as np
np1 = np.random.randint(0,10,6*5).reshape(6,5)

print(np1)

mm = np1.min(axis=1,keepdims=True) #(6,1) 연산 전 차원유지
MM = np1.max(axis=1,keepdims=True) #(6,1)

#mm = np1.min(axis=1)
#MM = np1.max(axis=1)
#mm = mm.reshape(6,-1) 방법2
#MM = MM.reshpae(6,-1) : mm이 갖고있는 크기 중 행6개를 만들고 뒤는 자동으로 채워라(-1)

np2 = (np1-mm)/(MM-mm)
np2
```

```python
#P50-0012
import numpy as np
import pandas as pd
csv_path = "https://raw.githubusercontent.com/ksoldier0419/DSet/main/ndarray/100_6_5.csv"
df1 = pd.read_csv(csv_path, header=None)
np1 = np.array(df1)

print(np1) #(6,5)

var1 = (np1**2).mean(axis=1,keepdims=True) - np1.mean(axis=1,keepdims=True)*2 #제평-평제
std1 = np.sqrt(var1) #(6,1)
print(std1.shape)

np2 = (np1-np1.mean(axis=1,keepdims=True))/std1
print(np2)
```

```python
#P50-0101
from sklearn import linear_model #사이킥런
import matplotlib.pyplot as plt

reg = linear_model.LinearRegression() #linear_model 안의 LinearRegression 함수를
#이용해서 reg 객체 생성

XX = [0,1,2]
YY = [3,3.5,5.5] #y=x+3

#plt.scatter(XX,YY)
#plt.show() #데이터 시각화 먼저 해보기.


XX = np.array(XX) #sklearn은 입력값이 무조건 2차원, 출력값은 상관X
XX = XX.reshape(-1,1) #T를 이용하지 않는다.
#학습시킨다 = fit
reg.fit(XX,YY)

print(reg.coef_)
print(reg.intercept_)
print(reg.score(XX,YY))
print(reg.predict([[5]])) #2차원 데이터 입력

plt.scatter(XX,YY)
pred = reg.predict(XX)
plt.plot(XX,pred, color='red', linewidth=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model

reg = linear_model.LinearRegression()

XX = [174,152,138,128,186]
YY = [71,55,46,38,88]

XX = np.array(XX)
XX = XX.reshape(-1,1)
reg.fit(XX,YY)

plt.scatter(XX,YY)
pred = reg.predict(XX)
plt.plot(XX,pred,color='green')

plt.scatter(160,reg.predict([[160]]),color='red')
plt.show()
```
