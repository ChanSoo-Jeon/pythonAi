```python
import numpy as np
my_vector = np.array([1,2,3,4,5,6])
selection = my_vector % 2 == 0
my_vector[selection]
# 넘파이 배열에 연산자를 적용하면 배열의 요소마다 연산자가 적용된다.
```

```python
first_matrix = np.array([[1,2,3],[4,5,6]])
second_matrix = np.array([1,2,3])
first_matrix + second_matrix
#넘파이 배열끼리 연산을 할 때, 크기가 다르면 넘파이는 자동으로 배열의 크기를 확장한다. = 브로드캐스팅 기능
```

```python
arr1 = np.ones((3,3))
arr1[1:-1,1:-1]=0
arr1
#[a:b:c,d:e:f] a,b : row c : step d,e : column f : step
```

```python
arr1 = np.zeros((5,5))
arr1[0::2,1::2] = 1
arr1[1::2,0::2] = 1
arr1
```

```python
arr1 = np.random.random((3,3))
mean = np.mean(arr1)
std = np.std(arr1)
norm = (arr1 - mean)/std
print(mean,std,norm)
```

```python
vector1 = np.arange(0,10)
vector1[5:9] = -vector1[5:9]
vector1
```

```python
arr1 = np.arange(0,9).reshape(3,3)
print("원본 배열: \n",arr1)
print("모든 요소의 합: ",arr1.sum())
print("각 열의 합: ",arr1.sum(0))
print("각 행의 합: ",arr1.sum(1))
```

```python
vector1 = np.array([4,5])
vector2 = np.array([7,10])
print("원본 벡터 : \n",vector1,"\n",vector2)
print("벡터의 내적: ",vector1 @ vector2)
```
