```python
from sklearn import datasets
iris = datasets.load_iris() #데이터 가져오기
#sklearn은 데이터를 동일한 형태로 가져온다

#iris 
iris.feature_names
#iris.data
#iris.data.shape #(150,4)
#iris.target
#iris.target_names #[0,1,2]

```

```python
from sklearn.model_selection import train_test_split


X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = \
train_test_split(X,y,test_size=0.2, random_state=4)
#test_size=0.2 : test data 20%사용의미
print(X_train.shape)
print(X_test.shape)

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=6) #홀수 권장
knn.fit(X_train, y_train) #data 차원일치시킨후 모델 훈련

y_pred = knn.predict(X_test) 

from sklearn import metrics 
scores = metrics.accuracy_score(y_test,y_pred) #정확도 측정
print(scores)

classes = {0:'setosa',1:'versicolor',3:'virginica'}

x_new = [[3,4,5,2],[5,4,2,2]] #새로운 데이터 입력 후 예측

y_predict = knn.predict(x_new)

print(classes[y_predict[0]])
print(classes[y_predict[1]])

```
